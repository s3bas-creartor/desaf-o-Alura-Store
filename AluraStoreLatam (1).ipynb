# 2. Ventas por categoría

```python
def analizar_ventas_tienda(df_tienda, nombre_tienda):
    """
    Analiza la cantidad de productos vendidos por categoría para una tienda específica
    y muestra las categorías más populares.

    Args:
        df_tienda (pd.DataFrame): El DataFrame con los datos de la tienda.
        nombre_tienda (str): El nombre de la tienda para fines de visualización.
    """
    print(f"\n--- Analizando Ventas para {nombre_tienda} ---")

    # ¡IMPORTANTE! Hemos corregido el nombre de la columna aquí
    # de 'Categoría del producto' a 'Categoría del Producto'
    if 'Categoría del Producto' not in df_tienda.columns:
        print(f"Error: La columna 'Categoría del Producto' no se encontró en los datos de {nombre_tienda}.")
        print("Esto no debería ocurrir con los nombres de columna que proporcionaste.")
        return

    # Contar la cantidad de productos vendidos por categoría
    # ¡IMPORTANTE! Hemos corregido el nombre de la columna aquí también
    ventas_por_categoria = df_tienda['Categoría del Producto'].value_counts()

    print(f"\n--- Cantidad de productos vendidos por Categoría en {nombre_tienda} ---")
    print(ventas_por_categoria)

    # Mostrar las categorías más populares (por defecto, las 3 principales)
    if not ventas_por_categoria.empty:
        print(f"\n--- Top 3 Categorías Populares en {nombre_tienda} ---")
        print(ventas_por_categoria.head(3))
    else:
        print(f"No hay datos de categorías para mostrar en {nombre_tienda}.")
    print("-" * 50)

# --- EJECUTAR EL ANÁLISIS PARA CADA TIENDA ---
# Asumiendo que 'tienda', 'tienda2', 'tienda3', 'tienda4' ya son DataFrames de pandas cargados.
# Si aún no los tienes cargados, asegúrate de ejecutar tus líneas de pd.read_csv(url) antes.

analizar_ventas_tienda(tienda, "Tienda 1")
analizar_ventas_tienda(tienda2, "Tienda 2")
analizar_ventas_tienda(tienda3, "Tienda 3")
analizar_ventas_tienda(tienda4, "Tienda 4")
```

# 3. Calificación promedio de la tienda


```python
def calcular_calificacion_promedio(df_tienda, nombre_tienda):
    """
    Calcula la calificación promedio de los clientes para una tienda específica.

    Args:
        df_tienda (pd.DataFrame): El DataFrame con los datos de la tienda.
        nombre_tienda (str): El nombre de la tienda para fines de visualización.
    """
    print(f"\n--- Calculando Calificación Promedio para {nombre_tienda} ---")

    # Verificar si la columna 'Calificación' existe
    if 'Calificación' not in df_tienda.columns:
        print(f"Error: La columna 'Calificación' no se encontró en los datos de {nombre_tienda}.")
        print("Asegúrate de que la columna se llame exactamente 'Calificación' (respetando mayúsculas y espacios).")
        return

    # Convertir la columna 'Calificación' a tipo numérico, manejando posibles errores
    # 'coerce' convertirá valores no numéricos a NaN (Not a Number)
    df_tienda['Calificación'] = pd.to_numeric(df_tienda['Calificación'], errors='coerce')

    # Eliminar filas con calificaciones no válidas (NaN) antes de calcular el promedio
    df_valid_ratings = df_tienda.dropna(subset=['Calificación'])

    # Calcular la calificación promedio
    if not df_valid_ratings.empty:
        calificacion_promedio = df_valid_ratings['Calificación'].mean()
        print(f"La calificación promedio de los clientes en {nombre_tienda} es: {calificacion_promedio:.2f}")
    else:
        print(f"No hay calificaciones válidas para calcular el promedio en {nombre_tienda}.")
    print("-" * 50)

# --- EJECUTAR EL CÁLCULO PARA CADA TIENDA ---
# Asumiendo que 'tienda', 'tienda2', 'tienda3', 'tienda4' ya son DataFrames de pandas cargados.
# Si aún no los tienes cargados, asegúrate de ejecutar tus líneas de pd.read_csv(url) antes.

calcular_calificacion_promedio(tienda, "Tienda 1")
calcular_calificacion_promedio(tienda2, "Tienda 2")
calcular_calificacion_promedio(tienda3, "Tienda 3")
calcular_calificacion_promedio(tienda4, "Tienda 4")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def analizar_y_visualizar_ventas_producto(df_tienda, nombre_tienda, top_n=5):
    """
    Identifica los productos más y menos vendidos y los visualiza,
    sin imprimir las tablas de texto directamente en la consola.

    Args:
        df_tienda (pd.DataFrame): El DataFrame con los datos de la tienda.
        nombre_tienda (str): El nombre de la tienda para fines de visualización.
        top_n (int): El número de productos Top N a mostrar (por defecto 5).
    """
    # Puedes comentar o eliminar esta línea si no quieres que aparezca
    # print(f"\n--- Analizando y Visualizando Productos Vendidos para {nombre_tienda} ---")

    if 'Producto' not in df_tienda.columns:
        print(f"Error: La columna 'Producto' no se encontró en los datos de {nombre_tienda}.")
        print("Asegúrate de que la columna se llame exactamente 'Producto'.")
        return

    conteo_productos = df_tienda['Producto'].value_counts()

    if conteo_productos.empty:
        print(f"No hay datos de productos para analizar en {nombre_tienda}.")
        return

    # --- Productos Más Vendidos ---
    productos_mas_vendidos = conteo_productos.head(top_n)
    # Comenta o elimina las siguientes líneas si no quieres las tablas impresas
    print(f"\n--- Top {top_n} Productos Más Vendidos en {nombre_tienda} ---")
    print(productos_mas_vendidos)

    # --- Productos Menos Vendidos ---
    productos_menos_vendidos = conteo_productos.tail(top_n)
    # Comenta o elimina las siguientes líneas si no quieres las tablas impresas
    print(f"\n--- Top {top_n} Productos Menos Vendidos en {nombre_tienda} ---")
    print(productos_menos_vendidos)

    # --- Visualización ---
    plt.figure(figsize=(12, 6))
    plt.suptitle(f'Análisis de Ventas de Productos en {nombre_tienda}', fontsize=16) # Título general para la figura

    # Gráfico de barras para productos más vendidos
    plt.subplot(1, 2, 1) # 1 fila, 2 columnas, primer gráfico
    # Corrección para FutureWarning: asignamos 'y' a 'hue' y 'legend=False'
    sns.barplot(x=productos_mas_vendidos.values, y=productos_mas_vendidos.index, palette='viridis', hue=productos_mas_vendidos.index, legend=False)
    plt.title(f'Top {top_n} Productos Más Vendidos') # Título específico del subplot
    plt.xlabel('Cantidad de Ventas')
    plt.ylabel('Producto')
    plt.grid(axis='x', linestyle='--', alpha=0.7)

    # Gráfico de barras para productos menos vendidos
    plt.subplot(1, 2, 2) # 1 fila, 2 columnas, segundo gráfico
    # Corrección para FutureWarning: asignamos 'y' a 'hue' y 'legend=False'
    sns.barplot(x=productos_menos_vendidos.values, y=productos_menos_vendidos.index, palette='mako', hue=productos_menos_vendidos.index, legend=False)
    plt.title(f'Top {top_n} Productos Menos Vendidos') # Título específico del subplot
    plt.xlabel('Cantidad de Ventas')
    plt.ylabel('Producto')
    plt.grid(axis='x', linestyle='--', alpha=0.7)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Ajusta el diseño, dejando espacio para el suptitle
    plt.show() # Muestra los gráficos
    # print("-" * 50) # Puedes comentar o eliminar esta línea también

# --- EJECUTAR EL ANÁLISIS Y VISUALIZACIÓN PARA CADA TIENDA ---
# Asumiendo que 'tienda', 'tienda2', 'tienda3', 'tienda4' ya son DataFrames de pandas cargados.
# Si aún no los tienes cargados, asegúrate de ejecutar tus líneas de pd.read_csv(url) antes.

analizar_y_visualizar_ventas_producto(tienda, "Tienda 1")
analizar_y_visualizar_ventas_producto(tienda2, "Tienda 2")
analizar_y_visualizar_ventas_producto(tienda3, "Tienda 3")
analizar_y_visualizar_ventas_producto(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
def calcular_costo_envio_promedio(df_tienda, nombre_tienda):
    """
    Calcula el costo de envío promedio de los productos para una tienda específica.

    Args:
        df_tienda (pd.DataFrame): El DataFrame con los datos de la tienda.
        nombre_tienda (str): El nombre de la tienda para fines de visualización.
    """
    print(f"\n--- Calculando Costo de Envío Promedio para {nombre_tienda} ---")

    # Verificar si la columna 'Costo de envío' existe.
    # Es crucial que esta columna tenga el nombre exacto.
    if 'Costo de envío' not in df_tienda.columns:
        print(f"Error: La columna 'Costo de envío' no se encontró en los datos de {nombre_tienda}.")
        print("Asegúrate de que la columna se llame exactamente 'Costo de envío' (respetando mayúsculas y espacios).")
        return

    # Convertir la columna 'Costo de envío' a tipo numérico, manejando posibles errores.
    # 'coerce' convertirá valores no numéricos a NaN (Not a Number).
    df_tienda['Costo de envío'] = pd.to_numeric(df_tienda['Costo de envío'], errors='coerce')

    # Eliminar filas con costos de envío no válidos (NaN) antes de calcular el promedio.
    df_valid_shipping_costs = df_tienda.dropna(subset=['Costo de envío'])

    # Calcular el costo de envío promedio.
    if not df_valid_shipping_costs.empty:
        costo_envio_promedio = df_valid_shipping_costs['Costo de envío'].mean()
        print(f"El costo de envío promedio por producto en {nombre_tienda} es: ${costo_envio_promedio:.2f}")
    else:
        print(f"No hay costos de envío válidos para calcular el promedio en {nombre_tienda}.")
    print("-" * 50)

# --- EJECUTAR EL CÁLCULO PARA CADA TIENDA ---
# Asumiendo que 'tienda', 'tienda2', 'tienda3', 'tienda4' ya son DataFrames de pandas cargados.
# Si aún no los tienes cargados, asegúrate de ejecutar tus líneas de pd.read_csv(url) antes.

calcular_costo_envio_promedio(tienda, "Tienda 1")
calcular_costo_envio_promedio(tienda2, "Tienda 2")
calcular_costo_envio_promedio(tienda3, "Tienda 3")
calcular_costo_envio_promedio(tienda4, "Tienda 4")
```

#Informe Final

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- FUNCIONES DE ANÁLISIS (Re-utilizamos las que ya creamos, con ligeras modificaciones para devolver valores) ---

def get_ventas_por_categoria(df_tienda):
    """Obtiene la serie de ventas por categoría."""
    if 'Categoría del Producto' not in df_tienda.columns:
        return pd.Series(dtype='int') # Retorna serie vacía si no existe la columna
    return df_tienda['Categoría del Producto'].value_counts()

def get_calificacion_promedio(df_tienda):
    """Calcula y retorna la calificación promedio."""
    if 'Calificación' not in df_tienda.columns:
        return None
    df_tienda['Calificación'] = pd.to_numeric(df_tienda['Calificación'], errors='coerce')
    df_valid_ratings = df_tienda.dropna(subset=['Calificación'])
    return df_valid_ratings['Calificación'].mean() if not df_valid_ratings.empty else None

def get_costo_envio_promedio(df_tienda):
    """Calcula y retorna el costo de envío promedio."""
    if 'Costo de envío' not in df_tienda.columns:
        return None
    df_tienda['Costo de envío'] = pd.to_numeric(df_tienda['Costo de envío'], errors='coerce')
    df_valid_shipping_costs = df_tienda.dropna(subset=['Costo de envío'])
    return df_valid_shipping_costs['Costo de envío'].mean() if not df_valid_shipping_costs.empty else None

# --- Asumimos que tus DataFrames ya están cargados ---
# Por ejemplo, si los cargaste así:
# tienda = pd.read_csv(url)
# tienda2 = pd.read_csv(url2)
# tienda3 = pd.read_csv(url3)
# tienda4 = pd.read_csv(url4)

# Y si fue necesario renombrar columnas, ya lo hiciste:
# if 'Category' in tienda.columns:
#     tienda.rename(columns={'Category': 'Categoría del Producto'}, inplace=True)
# etc.

# --- Recopilar datos clave para la visualización ---
tiendas_data = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

# Diccionarios para almacenar los resultados
calificaciones_promedio = {}
costos_envio_promedio = {}
top_categorias_por_tienda = {}

for nombre, df in tiendas_data.items():
    calificaciones_promedio[nombre] = get_calificacion_promedio(df)
    costos_envio_promedio[nombre] = get_costo_envio_promedio(df)
    # Obtenemos las top categorías para cada tienda
    top_categorias_por_tienda[nombre] = get_ventas_por_categoria(df).head(5) # Top 5 de cada tienda

# Convertir diccionarios a Series para fácil manejo con pandas
calificaciones_promedio_series = pd.Series(calificaciones_promedio)
costos_envio_promedio_series = pd.Series(costos_envio_promedio)

# --- CREACIÓN DE LOS GRÁFICOS ---

# Gráfico 1: Top 5 Categorías de Productos más Vendidas por Tienda (Gráfico de Barras)

print("\nGenerando gráficos de Top 5 Categorías de Productos por Tienda...")
for nombre_tienda, top_categorias in top_categorias_por_tienda.items():
    if not top_categorias.empty:
        plt.figure(figsize=(10, 6))
        sns.barplot(x=top_categorias.values, y=top_categorias.index, palette='viridis', hue=top_categorias.index, legend=False)
        plt.title(f'Top 5 Categorías de Productos más Vendidas en {nombre_tienda}', fontsize=14)
        plt.xlabel('Cantidad de Ventas', fontsize=12)
        plt.ylabel('Categoría del Producto', fontsize=12)
        plt.grid(axis='x', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()
    else:
        print(f"No hay datos de categorías para {nombre_tienda} para graficar.")


# Gráfico 2: Calificación Promedio de Clientes por Tienda (Gráfico de Barras Comparativas)

print("\nGenerando gráfico de Calificación Promedio de Clientes por Tienda...")
plt.figure(figsize=(10, 6))
sns.barplot(x=calificaciones_promedio_series.index, y=calificaciones_promedio_series.values, palette='magma')
plt.title('Calificación Promedio de Clientes por Tienda', fontsize=16)
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Calificación Promedio', fontsize=12)
plt.ylim(0, 5) # Asumiendo calificaciones de 1 a 5
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Añadir los valores exactos sobre las barras
for index, value in enumerate(calificaciones_promedio_series.values):
    if value is not None:
        plt.text(index, value + 0.05, f'{value:.2f}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()


# Gráfico 3: Costo de Envío Promedio vs. Calificación Promedio por Tienda (Gráfico de Dispersión)

print("\nGenerando gráfico de Dispersión: Costo de Envío Promedio vs. Calificación Promedio...")

# Asegurarse de que no haya Nones en los datos a graficar
df_comparativo = pd.DataFrame({
    'Costo de Envío Promedio': costos_envio_promedio_series,
    'Calificación Promedio': calificaciones_promedio_series
}).dropna() # Eliminar filas donde falte algún dato

if not df_comparativo.empty:
    plt.figure(figsize=(10, 7))
    sns.scatterplot(
        data=df_comparativo,
        x='Costo de Envío Promedio',
        y='Calificación Promedio',
        hue=df_comparativo.index, # Usar el nombre de la tienda para colorear los puntos
        s=200, # Tamaño de los puntos
        palette='tab10',
        legend='full'
    )
    plt.title('Relación entre Costo de Envío Promedio y Calificación Promedio', fontsize=16)
    plt.xlabel('Costo de Envío Promedio (COP)', fontsize=12)
    plt.ylabel('Calificación Promedio (1-5)', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)

    # Añadir etiquetas de texto para cada punto
    for i, row in df_comparativo.iterrows():
        plt.text(row['Costo de Envío Promedio'], row['Calificación Promedio'], i,
                 fontsize=10, ha='right', va='bottom')

    plt.tight_layout()
    plt.show()
else:
    print("No hay suficientes datos válidos para generar el gráfico de dispersión (Costo de Envío vs. Calificación).")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Asumimos que tus DataFrames ya están cargados ---
# Es fundamental que las variables 'tienda', 'tienda2', 'tienda3', 'tienda4'
# contengan tus DataFrames de pandas con las columnas 'lat' y 'lon'.

# Diccionario para agrupar tus tiendas, facilitando la iteración
tiendas_data = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

# --- 1. Visualización de la Distribución Geográfica de Ventas (Scatter Plots Individuales) ---
print("\n--- Generando gráficos de dispersión de la distribución geográfica de ventas por tienda ---")

for nombre_tienda, df_tienda in tiendas_data.items():
    plt.figure(figsize=(10, 8))

    # Asegurarse de que las columnas 'lat' y 'lon' existan
    if 'lat' not in df_tienda.columns or 'lon' not in df_tienda.columns:
        print(f"Advertencia: Las columnas 'lat' o 'lon' no se encontraron en los datos de {nombre_tienda}. Saltando el gráfico geográfico individual.")
        continue

    # Convertir a numérico (importante para evitar errores si los datos no son float/int)
    df_tienda['lat'] = pd.to_numeric(df_tienda['lat'], errors='coerce')
    df_tienda['lon'] = pd.to_numeric(df_tienda['lon'], errors='coerce')

    # Eliminar filas con valores nulos en lat/lon para la graficación
    df_plot = df_tienda.dropna(subset=['lat', 'lon'])

    if df_plot.empty:
        print(f"No hay datos de latitud/longitud válidos para {nombre_tienda} para graficar el scatter plot.")
        continue

    # Gráfico de dispersión: cada punto es una venta
    # 'alpha' ayuda a ver la densidad donde los puntos se superponen
    sns.scatterplot(x='lon', y='lat', data=df_plot, alpha=0.3, s=50, color='blue')

    plt.title(f'Distribución Geográfica de Ventas para {nombre_tienda}', fontsize=16)
    plt.xlabel('Longitud', fontsize=12)
    plt.ylabel('Latitud', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()


# --- 2. Visualización Consolidada de Ventas por Tienda (Scatter Plot Agrupado) ---
print("\n--- Generando gráfico de dispersión consolidado de ventas de todas las tiendas ---")

# Combinar todos los DataFrames en uno solo para el análisis consolidado
all_sales_data_consolidated = pd.DataFrame()
for nombre_tienda, df_tienda in tiendas_data.items():
    if 'lat' in df_tienda.columns and 'lon' in df_tienda.columns:
        df_temp = df_tienda[['lat', 'lon']].copy()
        df_temp['Tienda'] = nombre_tienda # Añadimos el nombre de la tienda para diferenciarlos por color
        df_temp['lat'] = pd.to_numeric(df_temp['lat'], errors='coerce')
        df_temp['lon'] = pd.to_numeric(df_temp['lon'], errors='coerce')
        df_temp.dropna(subset=['lat', 'lon'], inplace=True)
        all_sales_data_consolidated = pd.concat([all_sales_data_consolidated, df_temp])
    else:
        print(f"Advertencia: No se pueden incluir los datos geográficos de {nombre_tienda} en el gráfico de dispersión consolidado.")

if not all_sales_data_consolidated.empty:
    plt.figure(figsize=(12, 10))
    sns.scatterplot(
        x='lon',
        y='lat',
        data=all_sales_data_consolidated,
        hue='Tienda',  # Colorear los puntos por el nombre de la tienda
        alpha=0.6,     # Transparencia para ver la densidad subyacente
        s=50,          # Tamaño de los puntos
        palette='tab10', # Paleta de colores para las tiendas
        legend='full'
    )
    plt.title('Distribución Geográfica Consolidada de Ventas por Tienda', fontsize=18)
    plt.xlabel('Longitud', fontsize=14)
    plt.ylabel('Latitud', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
else:
    print("No hay suficientes datos geográficos válidos de todas las tiendas para generar un gráfico consolidado.")


# --- 3. Mapa de Calor de Densidad de Ventas (Consolidado de Todas las Tiendas) ---
print("\n--- Generando Mapa de Calor de Densidad de Ventas por Ubicación Geográfica (Consolidado) ---")

# Reutilizamos 'all_sales_data_consolidated' ya que contiene todos los datos limpios de lat/lon
if not all_sales_data_consolidated.empty:
    plt.figure(figsize=(12, 10))

    # Usamos sns.kdeplot para crear el mapa de calor de densidad
    sns.kdeplot(
        x='lon',
        y='lat',
        data=all_sales_data_consolidated,
        fill=True,       # Rellena las áreas con color
        cmap='viridis',  # Paleta de colores. Otras buenas opciones: 'magma', 'hot_r', 'YlOrRd'
        levels=20,       # Número de niveles de contorno para la suavidad del color
        alpha=0.8,       # Transparencia general del mapa de calor
        cbar=True,       # Muestra la barra de color (leyenda de densidad)
        cbar_kws={'label': 'Densidad de Ventas'} # Etiqueta para la barra de color
    )

    # Opcional: Puedes superponer los puntos individuales si quieres ver la distribución exacta de las ventas
    # sns.scatterplot(x='lon', y='lat', data=all_sales_data_consolidated, color='white', s=10, alpha=0.1)

    plt.title('Mapa de Calor: Densidad de Ventas por Ubicación Geográfica (Todas las Tiendas)', fontsize=18)
    plt.xlabel('Longitud', fontsize=14)
    plt.ylabel('Latitud', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
else:
    print("No hay suficientes datos geográficos válidos para generar el mapa de calor.")
```
